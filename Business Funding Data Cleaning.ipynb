## 1. Observations from Exploring the Data

While exploring the Business Funding Data, several important observations were made:

- The dataset contains multiple columns with missing values across both categorical and numerical variables.

- Some columns have a very high percentage of missing values, which reduces their usefulness for analysis.

- The dataset does not contain significant duplicate records, indicating that the entries are largely unique.

- A few numerical columns show large variations in values, which may influence statistical measures such as the mean.

- Some categorical columns contain repeated categories, making them suitable for imputation using the most frequent value.

These observations highlighted the need for careful handling of missing data and column relevance before performing further analysis.

## 2. Steps Taken to Clean, Preprocess, and Transform the Data

To prepare the dataset for analysis, the following preprocessing steps were applied:

### Handling Missing Values

- Missing values in categorical columns were replaced using the mode (most frequent value).

- Missing values in numerical columns were also filled using the mode, since the distribution suggested that the most common value better represents the data than the mean.

### Dropping Irrelevant Columns

- Columns with more than 70% missing values were removed from the dataset.

- These columns were considered not useful for meaningful analysis due to the large amount of missing information.

### Data Inspection

- The dataset was inspected to confirm that there were no duplicate rows.

- Data types and column structures were checked to ensure consistency.

After these preprocessing steps, the dataset became cleaner and more suitable for further exploratory data analysis and modeling.


## 3. Justification for the Techniques Used

The preprocessing techniques applied were chosen based on the nature of the dataset:

### Mode Imputation for Categorical Data
Since categorical variables cannot be averaged, the mode (most frequent value) is an appropriate way to replace missing entries while preserving the category distribution.

### Mode Imputation for Numerical Data
Some numerical columns contained outliers or extremely large values. Using the mean in such cases could distort the dataset. Therefore, using the mode helped maintain a more realistic representation of the data.

### Dropping Columns with Excessive Missing Values
Columns with over 70% missing data provide very limited information and may introduce noise into the analysis. Removing them improves overall data quality and reliability.

These decisions helped maintain data integrity while minimizing bias introduced during preprocessing.


## 4. Reflection on the Importance of Data Preprocessing

Data preprocessing is a critical step in real-world data analysis because raw data is often incomplete, inconsistent, or noisy.

Proper preprocessing ensures that:

- The dataset is accurate and reliable

- Analytical results are valid and meaningful

- Models built on the data produce better predictions and insights

Without proper preprocessing, analysis may lead to misleading conclusions or poor decision-making.

This exercise demonstrated how handling missing values, removing irrelevant columns, and understanding data structure are essential steps in transforming raw data into a usable dataset for analysis and visualization.
